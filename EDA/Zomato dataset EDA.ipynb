```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on = 'Country Code', how='left')
```

```python
final_df.head(2)
```

```python
## Check data types
final_df.dtypes
```

```python

final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
## pie chart - top 3 countries that uses zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct="%1.2f%%")
```

Observations:

 Zomato maximum transaction are from India
After that United State and then United Kingdom


```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating count'})
```

```python
ratings
```

## Observation
1. When rating is between 4.5 to 4.9 ---> are Excellent
2. When rating is between  4.0 to 3.4 --> are Very Good
3. When rating is between  3.5 to 3.9 --> are Good
4. When rating is between  2.5 to 2.9 --> are Average
5. When rating is between  2.0 to 2.4 --> are Poor

Maximum (2148) number of people didn't give ratings 


```python
sns.barplot(x="Aggregate rating", y="Rating count", hue='Rating color' ,data=ratings, palette=['blue','red','orange','yellow','green','green'])
```

## Observation
1. Not rated count is very high
2. Maximum num er of rating are between 2.5 to 3.4

```python
## Count plot - ploting w.r.t. categorical variables
sns.countplot(x="Rating color",data=ratings, palette=['blue','red','orange','yellow','green','green'])
```

Above is the frequncy, how frequntly our data is coming in dataset
